```r
distmat=read.table("dist_matrix.tab", header = TRUE, sep = "\t",row.names=1)
```

```r
distmat
```

```r
distmat['CP020543.1',]
distmat['CP024090.1',]
```

```r
njdistmat=nj(data.matrix(distmat))
```

```r
dat=fortify(njdistmat)
```

```r
dat
```

```r
dat$group<-'no'
```

```r
for (i in range(1,length(dat['node'])))
    if (dat[i,'label']=='CP020543.1')
        dat[i,'group']<-'yes'
        print('ok ')
    if (dat[i,'label']=='CP024090.1')
        dat[i,'group']<-'yes'
        print('ok ')
```

```r
pdf("mygraph2.pdf")
ggtree(dat,aes(x,y,color=group),size=0.2,  layout="fan")+geom_tiplab2(size=1,color="blue")
dev.off()
```
